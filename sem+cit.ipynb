```python
import json
import csv

# Step 1: Read the contents of train.txt
with open('train.txt', 'r', encoding='utf-8') as f:
    data = json.load(f)  # Assumes train.txt contains a valid JSON array

# Step 2: Prepare CSV writing
csv_file = "publications.csv"

fieldnames = [
    "publication_ID", "Citations", "pubDate", "language", "title", "journal",
    "abstract", "keywords", "authors", "venue", "doi"
]

with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    for entry in data:
        # Safely extract author names, handling missing 'name' keys
        author_names = [author.get('name', '') for author in entry.get("authors", [])]
        # Join author names, filtering out empty strings
        authors_string = "; ".join(name for name in author_names if name)

        writer.writerow({
            "publication_ID": entry.get("publication_ID"),
            "Citations": entry.get("Citations"),
            "pubDate": entry.get("pubDate"),
            "language": entry.get("language"),
            "title": entry.get("title"),
            "journal": entry.get("journal"),
            "abstract": entry.get("abstract"),
            "keywords": entry.get("keywords"),
            "authors": authors_string,  # Use the safely extracted author string
            "venue": entry.get("venue"),
            "doi": entry.get("doi")
        })

print(f"CSV file '{csv_file}' created successfully from 'train.txt'.")
```

```python
import pandas as pd

# 1. Load the CSV file
df = pd.read_csv('publications.csv')

# 2. Check initial columns
print("Columns available:", df.columns.tolist())

# 3. Drop any rows where title, abstract, or keywords are missing
df = df.dropna(subset=['title', 'abstract', 'keywords'])

# 4. Combine title, abstract, and keywords into one text field
def combine_text(row):
    return f"{row['title']}. {row['abstract']} Keywords: {row['keywords']}"

df['combined_text'] = df.apply(combine_text, axis=1)

# 5. Display a few examples to confirm
print("\nSample Combined Texts:")
print(df[['publication_ID', 'combined_text']].head())

# 6. Save cleaned version if needed (optional)
df.to_csv('cleaned_publications.csv', index=False)

# 7. Output: a DataFrame with 'combined_text' ready for embeddings
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd

# 1. Load the SentenceTransformer model
# (This one is small + fast + very good for semantic similarity)
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Load the cleaned publications.csv
df = pd.read_csv('cleaned_publications.csv')

# 3. Encode the combined text to get embeddings
texts = df['combined_text'].tolist()

print("Generating embeddings, this may take a few seconds...")
embeddings = model.encode(texts, show_progress_bar=True, convert_to_numpy=True)

# 4. Save embeddings (optional)
np.save('embeddings.npy', embeddings)  # saves as a numpy array
print("Embeddings saved as 'embeddings.npy'.")

# 5. Attach embeddings back to DataFrame (optional)
#    (Just for easy reference if needed later)
df['embedding'] = embeddings.tolist()

# 6. Save DataFrame with embeddings (optional)
df.to_pickle('publications_with_embeddings.pkl')
print("DataFrame with embeddings saved as 'publications_with_embeddings.pkl'.")
```

```python
import hdbscan
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# 1. Load embeddings and dataframe
embeddings = np.load('embeddings.npy')
df = pd.read_pickle('publications_with_embeddings.pkl')

# 2. Compute Cosine Similarity matrix
similarity_matrix = cosine_similarity(embeddings)

# 3. Convert similarity to distance
distance_matrix = 1 - similarity_matrix

# 4. Create HDBSCAN clusterer using precomputed distances
clusterer = hdbscan.HDBSCAN(
    min_cluster_size=30,
    metric='precomputed',
    cluster_selection_method='eom'
)

# 5. Fit the model
print("Clustering with cosine distance...")
cluster_labels = clusterer.fit_predict(distance_matrix)

# 6. Attach cluster labels back to DataFrame
df['cluster'] = cluster_labels

# 7. Save clustered data (optional)
df.to_pickle('publications_clustered.pkl')
print("Clustered data saved as 'publications_clustered.pkl'.")

# 8. Quick check: How many clusters formed?
n_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
print(f"\nNumber of clusters formed: {n_clusters}")

# 9. See a few examples
print(df[['publication_ID', 'title', 'cluster']].head(10))
```

```python
import networkx as nx
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
from tqdm import tqdm

# 1. Load clustered data
df = pd.read_pickle('publications_clustered_kmeans.pkl')
embeddings = np.load('embeddings.npy')

# 2. Prepare a dictionary to store paper rankings per cluster
cluster_paper_rankings = {}

# 3. Parameters
SIMILARITY_THRESHOLD = 0.7  # Only connect papers with similarity > 0.7 (can tune this)

# 4. Group by clusters
cluster_ids = df['cluster'].unique()
cluster_ids = [cid for cid in cluster_ids if cid != -1]  # Ignore noise (-1)

print(f"Building graphs and calculating PageRank for {len(cluster_ids)} clusters...")

for cluster_id in tqdm(cluster_ids):
    # Get all papers in this cluster
    cluster_df = df[df['cluster'] == cluster_id]
    idxs = cluster_df.index.tolist()

    # Extract their embeddings
    cluster_embeddings = embeddings[idxs]

    # Calculate cosine similarity between all pairs
    similarity_matrix = cosine_similarity(cluster_embeddings)

    # Create a graph
    G = nx.Graph()

    # Add nodes
    for idx in idxs:
        G.add_node(idx)

    # Add edges if similarity > threshold
    for i, idx_i in enumerate(idxs):
        for j, idx_j in enumerate(idxs):
            if i < j:  # Avoid double edges
                sim = similarity_matrix[i, j]
                if sim > SIMILARITY_THRESHOLD:
                    G.add_edge(idx_i, idx_j, weight=sim)

    # If graph is empty (very rare), skip
    if len(G.nodes) == 0:
        continue

    # Apply PageRank
    pagerank_scores = nx.pagerank(G, weight='weight')

    # Sort papers by PageRank
    sorted_papers = sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True)

    # Save top papers for this cluster
    cluster_paper_rankings[cluster_id] = sorted_papers

print("PageRank calculation done.")

# 5. Save PageRank rankings (optional)
import pickle
with open('pagerank_recommendations.pkl', 'wb') as f:
    pickle.dump(cluster_paper_rankings, f)

print("Saved PageRank recommendations as 'pagerank_recommendations.pkl'.")
```

```python
def recommend_papers(paper_id, df, cluster_paper_rankings, top_n=5):
    """
    Given a paper_id, recommend top_n similar papers from its cluster based on PageRank.
    """
    # Find cluster of the input paper
    paper_row = df[df['publication_ID'] == paper_id]
    if paper_row.empty:
        print(f"Paper ID {paper_id} not found.")
        return []

    cluster_id = paper_row['cluster'].values[0]

    # Get ranking list
    if cluster_id not in cluster_paper_rankings:
        print(f"No recommendations found for cluster {cluster_id}.")
        return []

    # Get top papers excluding the query paper itself
    ranking = [idx for idx, score in cluster_paper_rankings[cluster_id] if idx != paper_row.index[0]]

    # Get top_n papers
    top_indices = ranking[:top_n]
    recommended_papers = df.loc[top_indices][['publication_ID', 'title', 'abstract']]

    return recommended_papers
```

```python
# 1. Import necessary libraries
import pickle
import pandas as pd

# 2. Load your data
df = pd.read_pickle('publications_clustered_kmeans.pkl')

# 3. Load your PageRank recommendations
with open('pagerank_recommendations.pkl', 'rb') as f:
    cluster_paper_rankings = pickle.load(f)

# 5. Now call the function
paper_id = 12412787	  # <-- use any paper_id from your data
recommendations = recommend_papers(paper_id, df, cluster_paper_rankings, top_n=5)

# 6. Print recommendations
print(recommendations)
```

Build Citation Graph

```python
import pandas as pd
import networkx as nx

# 1. Load the original publications file
df = pd.read_csv('publications.csv')

# 2. Initialize a directed graph
citation_graph = nx.DiGraph()

# 3. Add nodes (each publication_ID becomes a node)
for pub_id in df['publication_ID']:
    citation_graph.add_node(pub_id)

# 4. Add edges based on citations
for _, row in df.iterrows():
    source_id = row['publication_ID']
    citations = row['Citations']

    if pd.isna(citations):
        continue  # No citations, skip

    cited_ids = citations.split(';')
    for target_id in cited_ids:
        try:
            target_id = int(target_id.strip())
            if target_id in df['publication_ID'].values:  # Only add if cited paper exists in dataset
                citation_graph.add_edge(source_id, target_id)
        except ValueError:
            continue  # Skip if malformed citation

# 5. Summary
print(f"Built citation graph with {citation_graph.number_of_nodes()} nodes and {citation_graph.number_of_edges()} edges.")

# 6. Save the graph (optional)
#nx.write_gpickle(citation_graph, 'citation_graph.gpickle')
#print("Saved citation graph as 'citation_graph.gpickle'.")
```

```python
import matplotlib.pyplot as plt

in_degrees = [citation_graph.in_degree(n) for n in citation_graph.nodes()]
out_degrees = [citation_graph.out_degree(n) for n in citation_graph.nodes()]

plt.hist(in_degrees, bins=50)
plt.title('In-degree (Number of times a paper is cited)')
plt.xlabel('Number of Citations')
plt.ylabel('Number of Papers')
plt.show()
```

Combine Semantic and Citation PageRank

```python
import pickle
import pandas as pd
import numpy as np

# 1. Load DataFrame
df = pd.read_pickle('publications_final_with_pagerank.pkl')

# 2. Load Citation PageRank
with open('citation_pagerank_scores.pkl', 'rb') as f:
    citation_pagerank = pickle.load(f)

# 3. Normalize semantic PageRank (already inside your df as 'pagerank_score')
semantic_scores = df['pagerank_score'].fillna(0).values
semantic_scores_normalized = (semantic_scores - np.min(semantic_scores)) / (np.max(semantic_scores) - np.min(semantic_scores))

# 4. Normalize citation PageRank
df['citation_pagerank'] = df['publication_ID'].map(citation_pagerank).fillna(0)
citation_scores = df['citation_pagerank'].values
citation_scores_normalized = (citation_scores - np.min(citation_scores)) / (np.max(citation_scores) - np.min(citation_scores))

# 5. Combine the two
ALPHA = 0.9  # semantic weight
BETA = 0.1   # citation weight

df['final_score'] = (ALPHA * semantic_scores_normalized) + (BETA * citation_scores_normalized)

# 6. Save final combined DataFrame
df.to_pickle('publications_final_combined.pkl')
print("Final combined scores saved in 'publications_final_combined.pkl'.")

# 7. See top papers
print("\nTop papers based on final score:")
print(df[['publication_ID', 'title', 'final_score']].sort_values('final_score', ascending=False).head(10))
```

```python
def recommend_papers_final(paper_id, df, top_n=5):
    """
    Given a paper_id, recommend top_n similar papers based on final combined score.
    """
    # Find cluster of the input paper
    paper_row = df[df['publication_ID'] == paper_id]
    if paper_row.empty:
        print(f"Paper ID {paper_id} not found.")
        return []

    cluster_id = paper_row['cluster'].values[0]

    # Get all papers in same cluster
    cluster_papers = df[df['cluster'] == cluster_id]

    # Sort by final_score
    sorted_papers = cluster_papers.sort_values('final_score', ascending=False)

    # Exclude the query paper itself
    sorted_papers = sorted_papers[sorted_papers['publication_ID'] != paper_id]

    return sorted_papers[['publication_ID', 'title', 'final_score']].head(top_n)
```

```python
recommendations = recommend_papers_final(paper_id=12412787, df=df, top_n=5)
print(recommendations)
```
