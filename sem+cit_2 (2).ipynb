```python
import networkx as nx
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
from tqdm import tqdm

# 1. Load clustered data
df = pd.read_pickle('publications_clustered_kmeans.pkl')
embeddings = np.load('embeddings.npy')

# 2. Prepare a dictionary to store paper rankings per cluster
cluster_paper_rankings = {}

# 3. Parameters
SIMILARITY_THRESHOLD = 0.7  # Only connect papers with similarity > 0.7 (can tune this)

# 4. Group by clusters
cluster_ids = df['cluster'].unique()
cluster_ids = [cid for cid in cluster_ids if cid != -1]  # Ignore noise (-1)

print(f"Building graphs and calculating PageRank for {len(cluster_ids)} clusters...")

for cluster_id in tqdm(cluster_ids):
    # Get all papers in this cluster
    cluster_df = df[df['cluster'] == cluster_id]
    idxs = cluster_df.index.tolist()

    # Extract their embeddings
    cluster_embeddings = embeddings[idxs]

    # Calculate cosine similarity between all pairs
    similarity_matrix = cosine_similarity(cluster_embeddings)

    # Create a graph
    G = nx.Graph()

    # Add nodes
    for idx in idxs:
        G.add_node(idx)

    # Add edges if similarity > threshold
    for i, idx_i in enumerate(idxs):
        for j, idx_j in enumerate(idxs):
            if i < j:  # Avoid double edges
                sim = similarity_matrix[i, j]
                if sim > SIMILARITY_THRESHOLD:
                    G.add_edge(idx_i, idx_j, weight=sim)

    # If graph is empty (very rare), skip
    if len(G.nodes) == 0:
        continue

    # Apply PageRank
    pagerank_scores = nx.pagerank(G, weight='weight')

    # Sort papers by PageRank
    sorted_papers = sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True)

    # Save top papers for this cluster
    cluster_paper_rankings[cluster_id] = sorted_papers

print("PageRank calculation done.")

# 5. Save PageRank rankings (optional)
import pickle
with open('pagerank_recommendations.pkl', 'wb') as f:
    pickle.dump(cluster_paper_rankings, f)

print("Saved PageRank recommendations as 'pagerank_recommendations.pkl'.")
```

```python
def recommend_papers(paper_id, df, cluster_paper_rankings, top_n=5):
    """
    Given a paper_id, recommend top_n similar papers from its cluster based on PageRank.
    """
    # Find cluster of the input paper
    paper_row = df[df['publication_ID'] == paper_id]
    if paper_row.empty:
        print(f"Paper ID {paper_id} not found.")
        return []

    cluster_id = paper_row['cluster'].values[0]

    # Get ranking list
    if cluster_id not in cluster_paper_rankings:
        print(f"No recommendations found for cluster {cluster_id}.")
        return []

    # Get top papers excluding the query paper itself
    ranking = [idx for idx, score in cluster_paper_rankings[cluster_id] if idx != paper_row.index[0]]

    # Get top_n papers
    top_indices = ranking[:top_n]
    recommended_papers = df.loc[top_indices][['publication_ID', 'title', 'abstract']]

    return recommended_papers
```

```python
import pandas as pd
import networkx as nx

# 1. Load the original publications file
df = pd.read_csv('publications.csv')

# 2. Initialize a directed graph
citation_graph = nx.DiGraph()

# 3. Add nodes (each publication_ID becomes a node)
for pub_id in df['publication_ID']:
    citation_graph.add_node(pub_id)

# 4. Add edges based on citations
for _, row in df.iterrows():
    source_id = row['publication_ID']
    citations = row['Citations']

    if pd.isna(citations):
        continue  # No citations, skip

    cited_ids = citations.split(';')
    for target_id in cited_ids:
        try:
            target_id = int(target_id.strip())
            if target_id in df['publication_ID'].values:  # Only add if cited paper exists in dataset
                citation_graph.add_edge(source_id, target_id)
        except ValueError:
            continue  # Skip if malformed citation

# 5. Summary
print(f"Built citation graph with {citation_graph.number_of_nodes()} nodes and {citation_graph.number_of_edges()} edges.")

# 6. Save the graph (optional)
#nx.write_gpickle(citation_graph, 'citation_graph.gpickle')
#print("Saved citation graph as 'citation_graph.gpickle'.")
```

```python
import pickle
import pandas as pd
import numpy as np

# 1. Load DataFrame
df = pd.read_pickle('publications_final_with_pagerank.pkl')

# 2. Load Citation PageRank
with open('citation_pagerank_scores.pkl', 'rb') as f:
    citation_pagerank = pickle.load(f)

# 3. Normalize semantic PageRank (already inside your df as 'pagerank_score')
semantic_scores = df['pagerank_score'].fillna(0).values
semantic_scores_normalized = (semantic_scores - np.min(semantic_scores)) / (np.max(semantic_scores) - np.min(semantic_scores))

# 4. Normalize citation PageRank
df['citation_pagerank'] = df['publication_ID'].map(citation_pagerank).fillna(0)
citation_scores = df['citation_pagerank'].values
citation_scores_normalized = (citation_scores - np.min(citation_scores)) / (np.max(citation_scores) - np.min(citation_scores))

# 5. Combine the two
ALPHA = 0.9  # semantic weight
BETA = 0.1   # citation weight

df['final_score'] = (ALPHA * semantic_scores_normalized) + (BETA * citation_scores_normalized)

# 6. Save final combined DataFrame
df.to_pickle('publications_final_combined.pkl')
print("Final combined scores saved in 'publications_final_combined.pkl'.")

# 7. See top papers
print("\nTop papers based on final score:")
print(df[['publication_ID', 'title', 'final_score']].sort_values('final_score', ascending=False).head(10))
```

```python
def recommend_papers_final(paper_id, df, top_n=5):
    """
    Given a paper_id, recommend top_n similar papers based on final combined score.
    """
    # Find cluster of the input paper
    paper_row = df[df['publication_ID'] == paper_id]
    if paper_row.empty:
        print(f"Paper ID {paper_id} not found.")
        return []

    cluster_id = paper_row['cluster'].values[0]

    # Get all papers in same cluster
    cluster_papers = df[df['cluster'] == cluster_id]

    # Sort by final_score
    sorted_papers = cluster_papers.sort_values('final_score', ascending=False)

    # Exclude the query paper itself
    sorted_papers = sorted_papers[sorted_papers['publication_ID'] != paper_id]

    return sorted_papers[['publication_ID', 'title', 'final_score']].head(top_n)
```

```python
import json
import pandas as pd

def load_and_prepare_txt(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    df = pd.DataFrame(data)

    # Drop rows missing key text fields
    df = df.dropna(subset=['title', 'abstract', 'keywords'])

    # Combine title, abstract, and keywords
    df['combined_text'] = df.apply(
        lambda row: f"{row['title']}. {row['abstract']} Keywords: {row['keywords']}",
        axis=1
    )

    # Convert citations from string to list
    def parse_citations(cite_entry):
        if pd.isna(cite_entry):
            return []
        if isinstance(cite_entry, list):
            return [int(cid) for cid in cite_entry if isinstance(cid, int) or (isinstance(cid, str) and cid.isdigit())]
        if isinstance(cite_entry, str):
            return [int(cid.strip()) for cid in cite_entry.split(';') if cid.strip().isdigit()]
        return []


    df['Citations'] = df['Citations'].apply(parse_citations)

    return df

# Load val and test
val_df = load_and_prepare_txt('val.txt')
test_df = load_and_prepare_txt('test.txt')

print("Validation and test data loaded.")
print(f"Val size: {len(val_df)} | Test size: {len(test_df)}")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import numpy as np
import pickle

# Load sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Load KMeans model used during training
with open('kmeans_model.pkl', 'rb') as f:
    kmeans = pickle.load(f)

# Load training DataFrame with cluster + final_score
train_df = pd.read_pickle('publications_final_combined.pkl')

# Load cluster recommendations: pre-sorted papers in each cluster
with open('pagerank_recommendations.pkl', 'rb') as f:
    semantic_ranking = pickle.load(f)  # idx → score (semantic part)

# Make a lookup: for each cluster, get top-N by final_score
def get_top_k_from_cluster(cluster_id, train_df, K=10):
    return (
        train_df[train_df['cluster'] == cluster_id]
        .sort_values('final_score', ascending=False)
        [['publication_ID']]
        .head(K)['publication_ID']
        .tolist()
    )

# Define function to embed + assign cluster + get top-K
def get_recommendations(df_eval, top_k=10):
    texts = df_eval['combined_text'].tolist()
    print("Generating embeddings for evaluation set...")
    embeddings = model.encode(texts, show_progress_bar=True, convert_to_numpy=True)

    print("Assigning clusters...")
    cluster_labels = kmeans.predict(embeddings)
    df_eval['cluster'] = cluster_labels

    print("Getting recommendations...")
    recommendations = []
    for i, row in df_eval.iterrows():
        cluster_id = row['cluster']
        top_k_ids = get_top_k_from_cluster(cluster_id, train_df, K=top_k)
        recommendations.append(top_k_ids)

    df_eval['recommended_ids'] = recommendations
    return df_eval

# Apply to validation and test sets
val_df = get_recommendations(val_df, top_k=10)
test_df = get_recommendations(test_df, top_k=10)

# Save results
val_df.to_pickle('val_with_recommendations.pkl')
test_df.to_pickle('test_with_recommendations.pkl')
print("Saved validation and test sets with recommendations.")
```

```python
def recall_at_k(true_ids, recommended_ids, k):
    if not true_ids:
        return np.nan  # Skip papers with no known citations
    hits = len(set(true_ids) & set(recommended_ids[:k]))
    return hits / len(true_ids)

def precision_at_k(true_ids, recommended_ids, k):
    if not recommended_ids:
        return np.nan
    hits = len(set(true_ids) & set(recommended_ids[:k]))
    return hits / k

def mrr_at_k(true_ids, recommended_ids, k):
    for i, rec_id in enumerate(recommended_ids[:k]):
        if rec_id in true_ids:
            return 1 / (i + 1)
    return 0
```

```python
def evaluate_recommendations(df, k=10):
    recall_scores = []
    precision_scores = []
    mrr_scores = []

    for _, row in df.iterrows():
        true = row['Citations']
        predicted = row['recommended_ids']

        recall_scores.append(recall_at_k(true, predicted, k))
        precision_scores.append(precision_at_k(true, predicted, k))
        mrr_scores.append(mrr_at_k(true, predicted, k))

    results = {
        f"Recall@{k}": np.nanmean(recall_scores),
        f"Precision@{k}": np.nanmean(precision_scores),
        f"MRR@{k}": np.nanmean(mrr_scores)
    }

    return results
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def average_embedding_similarity(df, query_embeddings, id_to_embedding, k=10):
    similarities = []

    for i, row in df.iterrows():
        query_vec = query_embeddings[i].reshape(1, -1)
        recommended_ids = row['recommended_ids'][:k]

        recommended_vecs = np.array([
            id_to_embedding.get(pid, np.zeros_like(query_vec)) for pid in recommended_ids
        ])

        sim_scores = cosine_similarity(query_vec, recommended_vecs)[0]
        if len(sim_scores) > 0:
            similarities.append(np.mean(sim_scores))

    return np.mean(similarities)
```

```python
def semantic_precision_at_k(df, k=10):
    precision_scores = []

    for _, row in df.iterrows():
        true_cluster = row['cluster']
        recommended_ids = row['recommended_ids'][:k]

        recommended_clusters = df[df['publication_ID'].isin(recommended_ids)]['cluster']
        matching = (recommended_clusters == true_cluster).sum()

        precision_scores.append(matching / k)

    return np.mean(precision_scores)
```

```python
# Semantic Precision@10 (how many recs are in same cluster)
def semantic_precision_at_k(df, train_df, k=10):
    scores = []
    for _, row in df.iterrows():
        true_cluster = row['cluster']
        rec_ids = row['recommended_ids']
        matched = train_df[train_df['publication_ID'].isin(rec_ids)]['cluster'] == true_cluster
        scores.append(matched.sum() / k)
    return np.mean(scores)

# Average embedding similarity
def average_embedding_similarity(df, test_embeddings, id_to_embedding, k=10):
    scores = []
    for i, row in df.iterrows():
        query_vec = test_embeddings[i].reshape(1, -1)
        rec_ids = row['recommended_ids']
        rec_vecs = np.array([id_to_embedding.get(rid, np.zeros(query_vec.shape[1])) for rid in rec_ids])
        sims = cosine_similarity(query_vec, rec_vecs)[0]
        if len(sims): scores.append(np.mean(sims))
    return np.mean(scores)

# Run both
precision_score = semantic_precision_at_k(test_df, train_df, k=10)
similarity_score = average_embedding_similarity(test_df, test_embeddings, embedding_lookup, k=10)

print(f"Semantic Precision@10: {precision_score:.4f}")
print(f"Average Cosine Similarity: {similarity_score:.4f}")
```

```python
import pandas as pd
import random
import json

# 1. Load the full CSV with citation info
df = pd.read_csv('publications.csv')

# 2. Drop rows missing required fields
df = df.dropna(subset=['title', 'abstract', 'keywords'])

# 3. Combine title, abstract, and keywords into one field
df['combined_text'] = df.apply(
    lambda row: f"{row['title']}. {row['abstract']} Keywords: {row['keywords']}",
    axis=1
)

# 4. Parse citations safely
def parse_citations(cite_entry):
    if pd.isna(cite_entry):
        return []
    if isinstance(cite_entry, list):
        return [int(cid) for cid in cite_entry if isinstance(cid, int) or (isinstance(cid, str) and cid.isdigit())]
    if isinstance(cite_entry, str):
        return [int(cid.strip()) for cid in cite_entry.split(';') if cid.strip().isdigit()]
    return []

df['Citations'] = df['Citations'].apply(parse_citations)

# 5. Create ID → text lookup
id_to_text = dict(zip(df['publication_ID'], df['combined_text']))

# 6. Filter citing papers
all_ids = set(df['publication_ID'])
train_citing_df = df[df['Citations'].apply(lambda x: isinstance(x, list) and len(x) > 0)]

triplets = []
print(f"Found {len(train_citing_df)} citing papers for triplet generation.")

for _, row in train_citing_df.iterrows():
    anchor_id = row['publication_ID']
    positives = [pid for pid in row['Citations'] if pid in id_to_text]

    if not positives:
        continue

    anchor_text = id_to_text[anchor_id]
    positive_text = id_to_text[random.choice(positives)]

    negative_id = random.choice(list(all_ids - set(positives) - {anchor_id}))
    negative_text = id_to_text[negative_id]

    triplets.append({
        "query": anchor_text,
        "positive": positive_text,
        "negative": negative_text
    })

# 7. Save triplets to file
with open("train_triplets.json", "w", encoding="utf-8") as f:
    json.dump(triplets, f, indent=2)

print(f"✅ Saved {len(triplets)} triplets to 'train_triplets.json'")
```

```python
from sentence_transformers import SentenceTransformer, InputExample, losses, models
from torch.utils.data import DataLoader
import json
import random

# 1. Load your triplets
with open("train_triplets.json", "r", encoding="utf-8") as f:
    triplets = json.load(f)

print(f"Loaded {len(triplets)} triplets.")

# 2. Convert to SentenceTransformer InputExample format
examples = []
for triplet in triplets:
    examples.append(InputExample(texts=[triplet["query"], triplet["positive"]]))

# 3. Define model (starting from pretrained MiniLM)
model = SentenceTransformer('all-MiniLM-L6-v2')  # You can switch to 'all-mpnet-base-v2' if you want higher quality

# 4. Set up data loader and loss
train_dataloader = DataLoader(examples, shuffle=True, batch_size=16)
train_loss = losses.MultipleNegativesRankingLoss(model)

# 5. Fine-tune
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=3,
    warmup_steps=100,
    output_path='fine_tuned_model'
)

print("✅ Fine-tuning complete. Model saved to 'fine_tuned_model'")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recall_at_k(true_ids, predicted_ids, k):
    if not true_ids: return np.nan
    return len(set(true_ids) & set(predicted_ids[:k])) / len(true_ids)

def precision_at_k(true_ids, predicted_ids, k):
    return len(set(true_ids) & set(predicted_ids[:k])) / k

def mrr_at_k(true_ids, predicted_ids, k):
    for i, pid in enumerate(predicted_ids[:k]):
        if pid in true_ids:
            return 1 / (i + 1)
    return 0

def evaluate_semantic(test_df, k=10):
    recall, precision, mrr = [], [], []

    for _, row in test_df.iterrows():
        recall.append(recall_at_k(row['Citations'], row['recommended_ids'], k))
        precision.append(precision_at_k(row['Citations'], row['recommended_ids'], k))
        mrr.append(mrr_at_k(row['Citations'], row['recommended_ids'], k))

    return {
        f"Recall@{k}": np.nanmean(recall),
        f"Precision@{k}": np.nanmean(precision),
        f"MRR@{k}": np.nanmean(mrr)
    }

results = evaluate_semantic(test_df, k=10)
print("FAISS-Based Global Recommendation Evaluation:")
for metric, value in results.items():
    print(f"{metric}: {value:.4f}")
```

```python
def average_embedding_similarity(test_df, test_embeddings, train_df, train_embeddings, k=10):
    embedding_lookup = dict(zip(train_df['publication_ID'], train_embeddings))
    similarities = []

    # Ensure index is aligned with embedding position
    test_df = test_df.reset_index(drop=True)

    for i, row in test_df.iterrows():
        query_vec = test_embeddings[i].reshape(1, -1)
        recommended_ids = row['recommended_ids'][:k]

        rec_vecs = np.array([
            embedding_lookup.get(pid, np.zeros(train_embeddings.shape[1])) for pid in recommended_ids
        ])

        sim_scores = cosine_similarity(query_vec, rec_vecs)[0]
        if len(sim_scores) > 0:
            similarities.append(np.mean(sim_scores))

    return np.mean(similarities)
```
